```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Read data from file
data = pd.read_csv("epa-sea-level.csv")
```

```python
X = data["Year"]  #independent variable/feature
y = data["CSIRO Adjusted Sea Level"]  #predictor
```

```python
# Create scatter plot
fig, ax = plt.subplots()

plt.scatter(X, y)

# Create first line of best fit
x = np.arange(1880, 2050, dtype=np.float64)
lr = linregress(X, y)
res = lr.slope * x + lr.intercept


plt.plot(x, res)
```

```python
# Create scatter plot
fig, ax = plt.subplots()

plt.scatter(X, y)

# Create first line of best fit
x = np.arange(1880, 2050, dtype=np.float64)
lr = linregress(X, y)
res = lr.slope * x + lr.intercept


plt.plot(x, res)

# Create second line of best fit
data_2 = data.iloc[120:][["Year", "CSIRO Adjusted Sea Level"]]

X_2 = data_2["Year"]
y_2 = data_2["CSIRO Adjusted Sea Level"]
x_2 = np.arange(2000, 2050)
lr2 = linregress(X_2, y_2)
res2 = lr2.slope * x_2 + lr2.intercept
plt.plot(x_2, res2)

# Add labels and title
plt.xticks(np.arange(1850, 2080, step=25, dtype=np.float64))
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.show()
```
